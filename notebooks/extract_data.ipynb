### Xử lí data

#### 1. Xử lí data từ json

```python
#Import thư viện cần thiết
import json
import pandas as pd
from itertools import zip_longest
```

```python
#Đọc file JSON
with open('../data/json/JSON1.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

rows = []

#Xử lý dữ liệu: tách dòng và ghép cặp
for item in data:
    src_id = item.get('id')
    # Lấy nội dung, mặc định là chuỗi rỗng nếu không có
    cn_text = item.get('cn', '')
    vi_text = item.get('vi', '')
    
    # Tách câu dựa trên xuống dòng
    cn_lines = cn_text.split('\n')
    vi_lines = vi_text.split('\n')
    
    # Ghép từng dòng tương ứng của 2 ngôn ngữ
    for cn, vi in zip_longest(cn_lines, vi_lines, fillvalue=''):
        rows.append({
            'src_id': src_id,
            'src_lang': cn,
            'tgt_lang': vi
        })

#Xuất ra file CSV
df = pd.DataFrame(rows)
df.to_csv('../data/preprocessing_data/data_converted.csv', index=False, encoding='utf-8-sig')
```
